```python
import numpy as np
import pandas as pd
import MelodyMachine as MM

%load_ext autoreload
%autoreload 1
%aimport MelodyMachine
```

```python
df = pd.read_csv('chord_db.csv')
df
```

```python
df['音程'] = df['度数'].apply(
    lambda inters: ','.
    join([str(MM.DEGREE_MAP[inter]) for inter in inters.split(',')])
    )
```

```python
df['组成音数'] = df['度数'].apply(
    lambda inters:
    len([MM.DEGREE_MAP[inter] for inter in inters.split(',')])
    )
```

```python
df = df[[
    '和弦标记',
    '名称',
    '优先级',
    '音程',
    '度数',
    '组成音数',
    '音程结构',
    '记号',
    '音高集合',
    ]]
```

```python
df.to_csv('chord_db.csv',index=False)
```
